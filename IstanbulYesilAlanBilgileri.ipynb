# 4.Kategorik Verilerin Kodlanması(Encoding)

Veri setinde kategorik verileri sayısal verilere dönüştürme işlemini yapacağız

```python
#hangileri kategorik inceliyoruz
df.dtypes
```

```python
#kategorik olanları tanımladık
categorical_columns = ['FAALİYET KONUSU', 'BİRİM', 2015, 2022, 2023]
```

```python
#Label Encoding:kategorik sütunları sayısal verilere dönüştürdük
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
for col in categorical_columns:
  df[col] = le.fit_transform(df[col].astype(str))
```

```python
#One-Hot Encoding:her kategori için ayrı bir sütun oluşturur
df = pd.get_dummies(df, columns=categorical_columns)
```

```python
print(df.columns)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

# 5.Veri Dönüşümü ve Eksik Veriler

a.Veri Dönüşümü(Transformatiion):
 Veri setinin analiz ve modelleme için uygun hale getirilmesini sağlar. Dönüşüm işlemleri veri setinin istatistiksel özelliklerini değiştirebilir ve bazı modellerin performansını artırabilir

### a.1.Ölçekleme(Scaling)
Verileri belirli bir ölçekte normalize eder.(Tüm sayısal özellikleri 0-1 aralığına sıkıştırmak gibi)

```python
print(df.to_string())
```

```python
from sklearn.preprocessing import StandardScaler
#2004 2005 2006 sütunlarındaki verileri standartlaştırır
scaler = StandardScaler()
df[[2004, 2005, 2006]] = scaler.fit_transform(df[[2004, 2005, 2006]])
```

```python
print(df.to_string())
```

```python
df.isnull().sum()
```

b)Outlier Tespiti(Outliers): Aykırı değerlerin verinin düzenini bozmaması için onları tespit eder ve işleriz

```python
#Z-Score Yöntemi:2004 2005 2006 sütunlarındaki aykırı değerleri tespit eder be temizler
from scipy import stats
import numpy as np
z_scores = np.abs(stats.zscore(df[[2004, 2005, 2006]]))
df_no_outliers = df[(z_scores < 3).all(axis=1)]
```

```python
print(df.to_string())
```

```python
from scipy import stats

z_scores = np.abs(stats.zscore(df[[2004, 2005, 2006, 2007]]))
df_no_outliers = df[(z_scores < 3).all(axis=1)]
```

```python
print(df.to_string())
```

```python
#temizlenmiş veri seti
print(df_no_outliers.describe())
```

```python
#bu veri setini kullanırım belki diye kaydediyorum
df_no_outliers.to_csv('cleaned_data.csv', index=False)
```

# 6.Discretization

```python
import pandas as pd

# Binleme işlemini yapın
df['Binned_2004'] = pd.cut(df[2004], bins=5, labels=['Bin1', 'Bin2', 'Bin3', 'Bin4', 'Bin5'])
```
